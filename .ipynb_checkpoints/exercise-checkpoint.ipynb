```python
import pandas as pd
import json

# Load the CSV file
food_df = pd.read_csv("Food_and_Nutrition_TEMPLATE.csv")

# Clean the disease names
food_df['Disease'] = food_df['Disease'].str.strip().str.lower()

# Convert to dictionary
food_dict = {}
for _, row in food_df.iterrows():
    disease = row['Disease']
    food_dict[disease] = {
        "Food to Eat": row['Food to Eat'],
        "Food to Avoid": row['Food to Avoid']
    }

# Save as JSON
with open("disease_food_map.json", "w") as f:
    json.dump(food_dict, f, indent=4)

print("✅ Food recommendations saved to disease_food_map.json")
```
